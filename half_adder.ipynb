## Write verilog

```python
%%writefile and.v
module and_gate(input wire in,in1,output wire out);
    assign out = in & in1;
endmodule
```

```python
%%writefile halfadder.v
module half_adder(input logic A, B,
                  output logic sum, c_Out);

    assign sum = A ^ B;
    assign c_Out = A & B;


endmodule
```
